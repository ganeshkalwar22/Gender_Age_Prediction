```python
from keras.layers import Dense,Conv2D,MaxPool2D,Flatten,Input,GlobalAveragePooling2D
from keras.models import Sequential
from keras import Model
from keras.applications.resnet50 import ResNet50
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
conv_base=ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3) )
```

```python
conv_base.trainable=False
```

```python
for layer in conv_base.layers:
    if layer.name.startswith("conv5"):
        layer.trainable = True
```

```python
for i in conv_base.layers:
    print(i.name)
```

```python
for i in conv_base.layers:
    print(i.trainable)
```

```python
x = conv_base.output
x = GlobalAveragePooling2D()(x)
x=Dense(512,activation='relu')(x)
x=Dense(256,activation='relu')(x)
x=Dense(64,activation='relu')(x)
```

```python
gender_output=Dense(1,activation='sigmoid',name="gender")(x)
age_output=Dense(1,activation='linear',name="age")(x)
```

```python
model=Model(inputs=conv_base.input,outputs=[age_output,gender_output])
```

```python
import os
import numpy as np
import cv2
from tensorflow.keras.utils import Sequence
```

```python
class UTKFaceGenerator(Sequence):
    def __init__(self, image_dir, batch_size=32):
        self.image_dir = image_dir
        self.batch_size = batch_size
        self.images = os.listdir(image_dir)

    def __len__(self):
        return len(self.images) // self.batch_size

    def __getitem__(self, idx):
        batch_files = self.images[idx*self.batch_size:(idx+1)*self.batch_size]

        X, age, gender = [], [], []

        for file in batch_files:
            try:
                a, g, _, _ = file.split("_", 3)
                img = cv2.imread(os.path.join(self.image_dir, file))
                img = cv2.resize(img, (200, 200))
                img = img / 255.0

                X.append(img)
                age.append(int(a))
                gender.append(int(g))

            except:
                continue

        return np.array(X), {
            "age": np.array(age),
            "gender": np.array(gender)
        }
```

```python
train_gen = UTKFaceGenerator(
    "/Users/ganeshkalwar/Desktop/Datasets/face_data/UTKFace",
    batch_size=8
)

val_gen = UTKFaceGenerator(
    "/Users/ganeshkalwar/Desktop/Datasets/face_data/UTKFace",
    batch_size=8
)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss={
        "age": "mae",
        "gender": "binary_crossentropy"
    },
    metrics={
        "age": "mae",
        "gender": "accuracy"
    }
)
```

```python
model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=1
)
```
